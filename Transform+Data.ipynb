```python
df_schools = pd.read_csv("Safe_Route_to_Schools_-_Priority_Schools.csv", sep=";")
```

```python
df_schools.head()
```

```python
df_schools = df_schools.dropna(subset=["Address"])
```

```python
df_schools["LATITUDE"] = np.full(len(df_schools), "NA", dtype=str)
df_schools["LONGITUDE"] = np.full(len(df_schools), "NA", dtype=str)
```

```python
from geopy.geocoders import Nominatim
from time import sleep
```

```python
geolocator = Nominatim(timeout=None)
```

```python
for idx, school in df_schools.iterrows():
    location = geolocator.geocode(school["Address"] + " " + school["Borough"] + " New York USA")
    sleep(1.5)
    if location != None:
        df_schools.set_value(idx, "LATITUDE", location.latitude)
        df_schools.set_value(idx, "LONGITUDE", location.longitude)    
    else:
        df_schools.set_value(idx, "LATITUDE", "nope")
        df_schools.set_value(idx, "LONGITUDE", "nope")   
    
```

```python
df_schools.head()
```

```python

df_schools['NEW_LATITUDE'] = [x.replace('.', ',') for x in df_schools['LATITUDE'].astype(str)]
df_schools['NEW_LONGITUDE'] = [x.replace('.', ',') for x in df_schools['LONGITUDE'].astype(str)]
```

```python
df_schools.head()
```

```python
df_schools.to_csv("schools.csv", sep=";", header=True)
```
